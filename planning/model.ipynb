## Choreography Planning



It assumes you have a dataset of dance moves in the specified format.

The dataset uses a proxy of pose landmarks of known movement vocabulary from 

The code below represents  a simple artificial intelligence using the K-Nearest Neighbors algorithm. The model learns patterns from the dataset and generates a new dance move based on the closest neighbors. The generated move will have similar duration and level properties as the neighbors in the dataset. The code only serves as a basic example and can be further improved by using accurate detection of landmarks and visual representation of the generated sequence.

* Movement vocabulary is based focused on movements found in Ballet, Graham based Modern, and Contemporary.
* This is an AI-based Minimum Viable Product (MVP) and the values of the landmarks are only using proxies.

## Summary

* **Problem Description**: How can we generate new moves based on detected landmarks (assuming these are detected inputs from a video)?
* **Area of AI**: Planning
* **Applied Algorithms**: Kinematic analysis, rotation
* **Results**: Generating new choreography

## Import Modules

```python
from data import movement_vocabulary
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import random
```

## KNN Model

```python
# Split the dataset into features (positions) and labels (duration and level)
features = [move["position"] for move in movement_vocabulary]
labels = [move["level"] for move in movement_vocabulary]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Train the kNN model
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

# Evaluate the model
score = knn.score(X_test, y_test)
print("Model score:", score)

# Generate a new dance move based on a given position
def generate_new_move_knn(position):
    predicted_values = knn.predict([position])[0]
    new_move = {
        "name": "new_move",
        "position": position,
        "level": round(predicted_values[1])
    }
    return new_move

# Generate a new position randomly (you can also provide a specific position)
new_position = {k: random.random() for k in features[0].keys()}
new_move = generate_new_move_knn(new_position)
print("Generated New Move:", new_move)
```
